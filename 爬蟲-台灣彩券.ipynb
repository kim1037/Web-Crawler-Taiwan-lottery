### 目標：透過網路爬蟲抓取台灣彩券期數、開獎號碼(依開獎順序、號碼大小兩種排序)及獎項金額

```python
import requests
from bs4 import BeautifulSoup

url='https://www.taiwanlottery.com.tw'
r= requests.get(url) #抓回原始碼
#使用BeautifulSoup解析
sp = BeautifulSoup(r.text,'lxml')
#找出威力彩的區塊
datas = sp.find('div',class_='contents_box02')  #find() >找到第一個tag
# print(datas)
#開獎期數的文字
title=datas.find('span', class_='font_black15').text
print('威力彩期數:',title)
numbers=datas.find_all('div', class_='ball_tx ball_green') #資料會用串列回傳

#把號碼印出來
print("開獎順序:",end="")
for i in range(0,6):
    print(numbers[i].text,end='')
print()

print("大小順序:",end="")    
for i in range(6,12):
    print(numbers[i].text,end='')
print()

num=datas.find('div', class_='ball_red') #資料會用串列回傳
print('第二區號碼:',num.text)   
```
